```python
from random import randint
```

```python
data = [randint(0,20) for _ in xrange(30)]
```

```python
data
```

```python
###構造統計詞頻的字典 c
c = dict.fromkeys(data, 0)
```

```python
c
```

```python
#統計詞頻
for x in data:
    c[x]+=1
```

```python
c
```

```python
### 使用 Counter 方便直接統計 出現頻度最高的元素
from collections import Counter
```

```python
c2 = Counter(data)
c2
```

```python
c2.most_common(3)
```

```python
#整體作為字符串讀入
txt = open('codingstyle').read()
```

```python
#使用 非单字字符 進行分隔
# \W 等價於 [^A-Za-z0-9_]
re.split('\W+', txt)
```

```python
c3 = Counter(re.split('\W+', txt))
```

```python
c3
```

```python
c3.most_common(10)
```
